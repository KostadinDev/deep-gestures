```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
def interpolate(entry):
    time = entry[:,0]
    x,y,z = entry[:,1], entry[:,2], entry[:,3]
    fx = interp1d(time,x,kind = 'cubic')
    fy = interp1d(time,y,kind = 'cubic')
    fz = interp1d(time,z,kind = 'cubic')
    new_time = np.linspace(time[0],time[-1], 150)
    interpolated_x = fx(new_time)
    interpolated_y = fy(new_time)
    interpolated_z = fz(new_time)
    interpolated_entry = [interpolated_x,interpolated_y,interpolated_z]
    pass
def throw_out(entry):
    pass
```

```python
def fix_entry(entry):
    if len(entry)<100:
        return interpolate(entry)
    elif len(entry)>100:
        return throw_out(entry)

    
```

```python
LR_data =[]
size = 0
with open('data/MAGICWAND_LRdash.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    entry = []
    for row in csv_reader:
        if row[0] == 'EXIT':
            LR_data.append(np.array(entry))
            size+=1
        elif row[0] == 'BEGIN':
            entry = []
        else:
            entry.append([float(i) for i in row])
print(f'Number of entries: {size}')
```

```python
O_data =[]
size = 0
with open('data/MAGICWAND_O.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    entry = []
    for row in csv_reader:
        if row[0] == 'EXIT':
            print(len(entry))
            O_data.append(np.array(entry))
            size+=1
        elif row[0] == 'BEGIN':
            entry = []
        else:
            entry.append([float(i) for i in row[:4]])
print(f'Number of entries: {size}')
```

```python
O_example = []
LR_example = []
for i in range(9):
    O_example.append(pd.DataFrame(O_data[i], columns= ['time','x','y','z']))
    LR_example.append(pd.DataFrame(LR_data[i], columns= ['time','x','y','z','xrot','yrot','zrot']))
```

```python
fig, ax = plt.subplots(3,2)
fig.set_figheight(15)
fig.set_figwidth(15)
fig.suptitle('O vs LR', fontsize=30)
counter = 0
for i in range(3):
    ax[i][0].plot(O_example[counter]['x'],O_example[counter]['z'], color='r')
    ax[i][1].plot(LR_example[counter]['y'],LR_example[counter]['z'])
    counter+=1
```

```python
fig = plt.figure(figsize = (20,10))
ax1 = fig.add_subplot(121, projection ='3d')
ax1.plot(O_example[counter]['x'], O_example[counter]['y'], O_example[counter]['z'])
ax1.set_title('(X,Y,Z) ACCELERATION')
ax1.view_init(30, 30)
ax2 = fig.add_subplot(122, projection ='3d')
ax2.plot(O_example[counter]['xrot'], O_example[counter]['yrot'], O_example[counter]['zrot'])
ax2.set_title('(X,Y,Z) ROTATIONAL ACCELERATION')
ax2.view_init(30, 30)
```
